```
sns.set_style('whitegrid')
sns.countplot(x = 'DIABETES' ,  data = df , hue = None)
```

```
sns.distplot( df['BMI'] , color = 'red')
```

```
fig,ax = plt.subplots(1 , 2 , figsize = (20,7))
sns.set_style('whitegrid')
sns.countplot(x = 'GENDER' ,  data = df , ax = ax[0])
df['SMOKING_HISTORY'].value_counts().plot.pie(autopct = "%1.1f%%" , ax = ax[1])
plt.show()
```

```
for i in range(0 , len(df)):
  if df['HYPERTENSION'].iloc[i] >  0:
    df['HYPERTENSION'].iloc[i] = "BP Patient"
  else:
    df['HYPERTENSION'].iloc[i] = "Non BP Patient"
```
